## An√°lisis de los RMSF de las trayectorias concatenadas de ambas replicas

### Variantes a 303K

```python
traj_WT_303 = md.load('MD_303/WT/WT_validated_mini_no_equi.xtc', top='MD_303/WT/1/WT_validated_mini.gro')
traj_S214H_303 = md.load('MD_303/S214H/S214H_validated_mini_no_equi.xtc', top='MD_303/S214H/1/S214H_validated_mini.gro')
traj_TSPETase_303 = md.load('MD_303/TSPETase/TSPETase_validated_mini_no_equi.xtc', top='MD_303/TSPETase/1/TSPETase_validated_mini.gro')
traj_DuraPETase_303 = md.load('MD_303/DuraPETase/DuraPETase_validated_mini_no_equi.xtc', top='MD_303/DuraPETase/1/DuraPETase_validated_mini.gro')
```

Se realiza una selecci√≥n de √∫nicamente los carbonos alpha para calcular el rmsf

```python
protein_sel_WT= traj_WT_303.topology.select('protein and name CA')
protein_traj_WT = traj_WT_303.atom_slice(protein_sel_WT)
protein_sel_S214H= traj_S214H_303.topology.select('protein and name CA')
protein_traj_S214H = traj_S214H_303.atom_slice(protein_sel_S214H)
protein_sel_TSPETase= traj_TSPETase_303.topology.select('protein and name CA')
protein_traj_TSPETase = traj_TSPETase_303.atom_slice(protein_sel_TSPETase)
protein_sel_DuraPETase= traj_DuraPETase_303.topology.select('protein and name CA')
protein_traj_DuraPETase = traj_DuraPETase_303.atom_slice(protein_sel_DuraPETase)
```

C√°lculo de los rmsf de las diferentes variantes

```python
## A 303K

rmsf_WT_303 = md.rmsf(protein_traj_WT,protein_traj_WT,0)
rmsf_S214H_303 = md.rmsf(protein_traj_S214H,protein_traj_S214H,0)
rmsf_TSPETase_303 = md.rmsf(protein_traj_TSPETase,protein_traj_TSPETase,0)
rmsf_DuraPETase_303 = md.rmsf(protein_traj_DuraPETase,protein_traj_DuraPETase,0)
```
