```python
traj_WT_1.center_coordinates()
rmsd_WT_1=md.rmsd(traj_WT_1,traj_WT_1,0)
traj_WT_1.center_coordinates()
rmsd_WT_2=md.rmsd(traj_WT_2,traj_WT_2,0)
traj_S214H_1.center_coordinates()
rmsd_S214H_1=md.rmsd(traj_S214H_1,traj_S214H_1,0)
traj_S214H_2.center_coordinates()
rmsd_S214H_2=md.rmsd(traj_S214H_2,traj_S214H_2,0)
traj_TSPETase_1.center_coordinates()
rmsd_TSPETase_1=md.rmsd(traj_TSPETase_1,traj_TSPETase_1,0)
traj_TSPETase_2.center_coordinates()
rmsd_TSPETase_2=md.rmsd(traj_TSPETase_2,traj_TSPETase_2,0)
traj_DuraPETase_1.center_coordinates()
rmsd_DuraPETase_1=md.rmsd(traj_DuraPETase_1,traj_DuraPETase_1,0)
traj_DuraPETase_2.center_coordinates()
rmsd_DuraPETase_2=md.rmsd(traj_DuraPETase_2,traj_DuraPETase_2,0)
```

```python
import decimal

def float_range(start, stop, step):
    while start < stop:
        yield float(start)
        start += decimal.Decimal(step)
```

```python
fig = plt.figure(figsize=(16,7))

plt.rcParams["font.family"] = "Times New Roman"

sim_time=list(float_range(0,25.001, 0.01))
plt.plot(sim_time, rmsd_WT_1, color='blue', label='WT')
plt.plot(sim_time, rmsd_WT_2, color='blue', label='WT')
plt.plot(sim_time, rmsd_S214H_1, color='yellow', label='S214H')
plt.plot(sim_time, rmsd_S214H_2, color='yellow', label='S214H')
plt.plot(sim_time, rmsd_TSPETase_1, color='green', label='TSPETase')
plt.plot(sim_time, rmsd_TSPETase_2, color='green', label='TSPETase')
plt.plot(sim_time, rmsd_DuraPETase_1, color='red', label='DuraPETase')
plt.plot(sim_time, rmsd_DuraPETase_2, color='red', label='DuraPETase')
plt.xlabel('Simulation time (ns)', size=40)
plt.ylabel('RMSD ()', size=35)
plt.xticks(size=20)
plt.yticks(size=20)
plt.ylim(0,0.6)
plt.xlim(0,25.5)
## Make the legend
WT = plt.Rectangle((0,0),1,1,fc="blue", edgecolor = 'none', alpha=.8)
S214H = plt.Rectangle((0,0),1,1,fc='yellow',  edgecolor = 'none', alpha=.8)
TS = plt.Rectangle((0,0),1,1,fc='green',  edgecolor = 'none', alpha=.8)
DP = plt.Rectangle((0,0),1,1,fc='red',  edgecolor = 'none', alpha=.8)
l = plt.legend([WT, S214H, TS, DP], ['WT', 'S214H', 'TSPETase', 'DuraPETase'], loc=2, ncol = 1, prop={'size':14})
l.draw_frame(False)

## Save the picture
plt.savefig('RMSD_withPET_303.jpg')

plt.show()
```

Se realiza una seleccin de nicamente los carbonos alpha para calcular el rmsf

```python
protein_sel_WT= traj_WT_303.topology.select('protein and name CA')
protein_traj_WT = traj_WT_303.atom_slice(protein_sel_WT)
protein_sel_S214H= traj_S214H_303.topology.select('protein and name CA')
protein_traj_S214H = traj_S214H_303.atom_slice(protein_sel_S214H)
protein_sel_TSPETase= traj_TSPETase_303.topology.select('protein and name CA')
protein_traj_TSPETase = traj_TSPETase_303.atom_slice(protein_sel_TSPETase)
protein_sel_DuraPETase= traj_DuraPETase_303.topology.select('protein and name CA')
protein_traj_DuraPETase = traj_DuraPETase_303.atom_slice(protein_sel_DuraPETase)
```

```python
## A 303K
rmsf_WT_303 = md.rmsf(protein_traj_WT,protein_traj_WT,0)
rmsf_S214H_303 = md.rmsf(protein_traj_S214H,protein_traj_S214H,0)
rmsf_TSPETase_303 = md.rmsf(protein_traj_TSPETase,protein_traj_TSPETase,0)
rmsf_DuraPETase_303 = md.rmsf(protein_traj_DuraPETase,protein_traj_DuraPETase,0)
```

### Clase/Objeto para poder hacer grficos agrupados de las caractersticas

```python
class SeabornFig2Grid():

    def __init__(self, seaborngrid, fig,  subplot_spec):
        self.fig = fig
        self.sg = seaborngrid
        self.subplot = subplot_spec
        if isinstance(self.sg, sns.axisgrid.FacetGrid) or \
            isinstance(self.sg, sns.axisgrid.PairGrid):
            self._movegrid()
        elif isinstance(self.sg, sns.axisgrid.JointGrid):
            self._movejointgrid()
        self._finalize()

    def _movegrid(self):
        """ Move PairGrid or Facetgrid """
        self._resize()
        n = self.sg.axes.shape[0]
        m = self.sg.axes.shape[1]
        self.subgrid = gridspec.GridSpecFromSubplotSpec(n,m, subplot_spec=self.subplot)
        for i in range(n):
            for j in range(m):
                self._moveaxes(self.sg.axes[i,j], self.subgrid[i,j])

    def _movejointgrid(self):
        """ Move Jointgrid """
        h= self.sg.ax_joint.get_position().height
        h2= self.sg.ax_marg_x.get_position().height
        r = int(np.round(h/h2))
        self._resize()
        self.subgrid = gridspec.GridSpecFromSubplotSpec(r+1,r+1, subplot_spec=self.subplot)

        self._moveaxes(self.sg.ax_joint, self.subgrid[1:, :-1])
        self._moveaxes(self.sg.ax_marg_x, self.subgrid[0, :-1])
        self._moveaxes(self.sg.ax_marg_y, self.subgrid[1:, -1])

    def _moveaxes(self, ax, gs):
        #https://stackoverflow.com/a/46906599/4124317
        ax.remove()
        ax.figure=self.fig
        self.fig.axes.append(ax)
        self.fig.add_axes(ax)
        ax._subplotspec = gs
        ax.set_position(gs.get_position(self.fig))
        ax.set_subplotspec(gs)

    def _finalize(self):
        plt.close(self.sg.fig)
        self.fig.canvas.mpl_connect("resize_event", self._resize)
        self.fig.canvas.draw()

    def _resize(self, evt=None):
        self.sg.fig.set_size_inches(self.fig.get_size_inches())
```
